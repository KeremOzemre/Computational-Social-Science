# Importing and storing citation and author in one array

```python
import pandas as pd
import numpy as np
from pathlib import Path


PATH = Path("CCS Authors dataset.csv")
df = pd.read_csv(PATH)
```

```python
# Cleaning
df.columns = [c.strip() for c in df.columns]
```

```python
# sørg for numeric citations
df["cited_by_count"] = pd.to_numeric(
    df["cited_by_count"], errors="coerce"
).fillna(0).astype(int)
```

```python
# Array with name and citations
author_citations = df.groupby("display_name")["cited_by_count"].sum()
```

```python
authors = author_citations.index.to_numpy()
citations = author_citations.to_numpy()
```

```python
# Tjek om det passer
print(df.columns)              
print(authors.shape, citations.shape)
print(authors[:3], citations[:3])
```

# Visualization with plotting

```python
min(citations), max(citations)
```

```python
xmin = min(citations)
xmax = max(citations)
```

```python
bins = np.linspace(xmin, xmax, 50)
```

```python
hist, edges = np.histogram(citations, bins=bins)
```

```python
x = (edges[1:] + edges[:-1]) / 2
```

```python
# Plot counts
fig, ax = plt.subplots()

ax.plot(x, hist, marker='.')

ax.set_xlabel('citations per author')
ax.set_ylabel('counts')

plt.show()
```

```python
# Density plot
hist, edges = np.histogram(citations, bins=bins, density=True)
x = (edges[1:] + edges[:-1]) / 2

fig, ax = plt.subplots()

ax.plot(x, hist, marker='.')

ax.set_xlabel('citations per author')
ax.set_ylabel('probability density')

plt.show()
```
