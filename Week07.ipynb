### lexical diversity:

Lexical diversity describes how varied the vocabulary in a text is.
It is usually measured as the ratio between the number of unique words and the total number of words.
A high lexical diversity means that the text uses many different words, while a low lexical diversity means that the same words are repeated often.

```python
# this value is the lexical diversity of text 1
def lexical_diversity(text):
    return len(set(text)) / len(text)

lexical_diversity(text1)
```

```python
import pandas as pd
import ast
import numpy as np

# load data
df = pd.read_csv("abstracts.csv")

# sikker konvertering fra string til dictionary
def safe_parse_index(x):
    if pd.isna(x):          # håndterer NaN
        return {}
    if isinstance(x, dict): # hvis den allerede er dict
        return x
    if isinstance(x, str):
        x = x.strip()
        if x == "":
            return {}
        try:
            return ast.literal_eval(x)
        except:
            return {}
    return {}

df["abstract_index"] = df["abstract_index"].apply(safe_parse_index)
```

```python
def inverted_index_to_text(index_dict):
    if not index_dict:   # tom dict
        return ""
    
    # find højeste position
    max_pos = max(pos for positions in index_dict.values() for pos in positions)
    
    # opret liste til ordene
    words = [""] * (max_pos + 1)
    
    # sæt hvert ord ind på de positioner det har
    for word, positions in index_dict.items():
        for pos in positions:
            words[pos] = word
    
    # fjern evt. tomme pladser og join til tekst
    return " ".join(word for word in words if word != "")
```

```python
df["text"] = df["abstract_index"].apply(inverted_index_to_text)
```

```python
df[["title", "text"]].head()
```

```python
import re

def clean_tokenize(text):
    if not isinstance(text, str) or text.strip() == "":
        return []
    
    # lowercase
    text = text.lower()
    
    
    text = re.sub(r'http\S+|www\S+', '', text)
    
    
    text = re.sub(r'[^a-z\s]', '', text)


    
    tokens = text.split()
    
    return tokens
```

```python
df["tokens"] = df["text"].apply(clean_tokenize)
```

```python
df[["title", "tokens"]].head()
```

```python
from collections import Counter

word_freq = Counter(all_tokens)

top_10 = word_freq.most_common(10)

top_10
```

```python

import matplotlib.pyplot as plt

def plot_frequency_rank(tokens):
    
    word_counts = Counter(tokens)
    
    
    frequencies = sorted(word_counts.values(), reverse=True)
    
    #rankss
    ranks = range(1, len(frequencies) + 1)
    
  
    plt.figure(figsize=(8, 5))
    plt.plot(ranks, frequencies)
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Word Rank (log scale)")
    plt.ylabel("Word Frequency (log scale)")
    plt.title("Frequency-Rank Plot (Zipf's Law)")
    plt.grid(True)
    plt.show()
```

```python
plot_frequency_rank(all_tokens)
```
