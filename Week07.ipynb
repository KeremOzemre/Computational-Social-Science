```python
import pandas as pd
import ast

df1 = pd.read_csv('abstracts.csv')
```

```python
s = df1["abstract_index"]
```

```python
def construct_abstract_from_inverted(inverted_abstract):
    d = ast.literal_eval(inverted_abstract)

    # Create position → word mapping
    position_to_word = {}
    for word, positions in d.items():
        for pos in positions:
            position_to_word[pos] = word
    
    

    abstract = [word for _, word in sorted(position_to_word.items())]
    return " ".join(abstract)


df1["text"] = s.dropna().apply(construct_abstract_from_inverted)
```

```python
import nltk
import re
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))


def tokenize(abstract):
    text = abstract.lower()
    
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    stems = [stemmer.stem(word) for word in filtered_words]
    
    
    return stems

df1["tokens"] = df1["text"].dropna().apply(tokenize)

```

```python
import numpy as np
from collections import Counter
from scipy.stats import chi2
import nltk
import math

clean_tokens = [
    t for t in all_tokens
    if t is not None and not (isinstance(t, float) and math.isnan(t))
]

bigrams = list(nltk.bigrams(clean_tokens))


bigram_counts = Counter(bigrams)
w1_counts = Counter([w1 for w1, _ in bigrams])
w2_counts = Counter([w2 for _, w2 in bigrams])
N = len(bigrams)

contingency_tables = {}
expected_tables = {}
chi2_scores = {}
p_values = {}
co_occurences = {}

for (w1, w2), nii in bigram_counts.items():
    # Observed values
    nio = w1_counts[w1] - nii
    noi = w2_counts[w2] - nii
    noo = N - (nii + nio + noi)

    observed = np.array([[nii, nio],
                         [noi, noo]])

    # Expected values
    row_sums = observed.sum(axis=1)
    col_sums = observed.sum(axis=0)
    expected = np.outer(row_sums, col_sums) / N

    # Chi-squared statistic
    chi2_stat = ((observed - expected) ** 2 / expected).sum()

    # p-value (df = 1)
    p_val = chi2.sf(chi2_stat, df=1)

    # Store results
    co_occurences[(w1, w2)] = nii
    contingency_tables[(w1, w2)] = observed
    expected_tables[(w1, w2)] = expected
    chi2_scores[(w1, w2)] = chi2_stat
    p_values[(w1, w2)] = p_val


    
```

```python
filtered_bigrams = {
    bg:(co_occurences[bg]) for bg in bigram_counts
    if (p_values[bg] < 0.001) and (co_occurences[bg] > 50)
}
```

```python
sorted_bigrams = sorted(
    filtered_bigrams.items(),
    key=lambda x: x[1],
    reverse=True
)
```

```python
import nltk
import re
nltk.download('punkt_tab')
from nltk.tokenize import MWETokenizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))
collocations = list(filtered_bigrams.keys())

mwe_tokenizer = MWETokenizer(collocations, separator='_')


def tokenize(abstract):
    text = abstract.lower()
    
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    words = text.split()
    
    # remove stopwords
    filtered_words = [word for word in words if word not in stop_words]
    
    # stemming
    stems = [stemmer.stem(word) for word in filtered_words]
    
    # apply collocation merging
    tokens = mwe_tokenizer.tokenize(stems)
    
    return tokens

df1["tokens"] = df1["text"].dropna().apply(tokenize)
```
