```python
import pandas as pd
import ast
```

```python
df1 = pd.read_csv('authors_data.csv')
df2 = pd.read_csv('works_ids.csv')
df3 = pd.read_csv('works_titles_abstracts.csv')
df4 = pd.read_csv('co_authors_data.csv')
df5 = pd.read_csv('co_authors_works.csv')
df6 = pd.read_csv('co_authors_works_titles_abstracts.csv')
```

```python
authors_dataset = pd.concat([df1, df4], ignore_index=True)
papers_dataset = pd.concat([df2,df5], ignore_index = True)
abstracts_dataset = pd.concat([df3,df6],ignore_index = True)
```

```python
authors_dataset = authors_dataset.drop(columns=['searched_name'])
authors_dataset = authors_dataset.drop_duplicates(subset = "author_id")
authors_dataset.to_csv('authors_dataset.csv', index=False)
```

```python
papers_dataset = papers_dataset.drop_duplicates(subset = "id")
papers_dataset.to_csv('papers_dataset.csv',index = False)
```

```python
papers_dataset["author_ids"] = papers_dataset["author_ids"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)
several_authors_paper_dataset = papers_dataset[
    papers_dataset["author_ids"].str.len() >= 2
]
several_authors_paper_dataset.to_csv('more_than_one_author_papers_dataset.csv',index = False)
```

```python
abstracts_dataset = abstracts_dataset.drop_duplicates(subset = "id")
abstracts_dataset.to_csv('abstracts_dataset.csv',index = False)
```
