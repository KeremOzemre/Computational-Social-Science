```python
import pandas as pd
import networkx as nx
import ast
```

```python
authors = pd.read_csv("../authors_dataset_clean.csv")
papers = pd.read_csv("../more_than_one_author_papers_dataset_clean.csv")
```

```python
G = nx.Graph()

for _, row in papers.iterrows():

    author_list = ast.literal_eval(row["author_ids"])

    for i in range(len(author_list)):
        for j in range(i+1, len(author_list)):
            G.add_edge(author_list[i], author_list[j])
```

```python
country_dict = dict(zip(authors["author_id"], authors["country_code"]))

nx.set_node_attributes(G, country_dict, "country")
```
