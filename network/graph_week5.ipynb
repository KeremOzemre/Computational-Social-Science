```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.random.normal(loc = 0,scale = 4,size = 10000)
distribution_mean = np.mean(x)
distribution_median = np.median(x)
```

```python
plt.hist(x, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Array")
plt.show()
```

```python
cum_average = []

for i in range(1,len(x)):
    subset = x[:i]
    cum_average.append(np.mean(subset))

cum_average = np.array(cum_average)
print(len(cum_average))


```

```python
std_average = []

for i in range(1, len(x)):
    subset = x[:i]
    std_average.append(np.std(subset))

cum_std = np.array(std_average)
len(cum_std)
```

```python
x_vals = np.arange(len(cum_average))
yerr = cum_std / np.sqrt(np.arange(1, len(cum_std)+1))

plt.plot(x_vals, cum_average)

plt.axhline(y=0, color="r", linestyle="--", label="mean = 0")

plt.errorbar(
    x_vals,
    cum_average,
    yerr=yerr,
    capsize=1,
    fmt="none",        # prevents double plotting
    ecolor="purple"
)

plt.xlabel("Iteration")
plt.ylabel("Cumulative Average")
plt.legend()
plt.show()
```

```python
cumulative_median = []

for i in range(1, len(x)):
    subset = x[:i]
    cumulative_median.append(np.median(subset))

cumulative_median = np.array(cumulative_median)
```

```python
plt.plot(x_vals, cumulative_median)

plt.axhline(y=0, color="r", linestyle="--", label=f"median = 0")

plt.xlabel("Iteration")
plt.ylabel("Cumulative Average")
plt.legend()
plt.show()
```

```python
a = 0.5
x_m = 1
s = (np.random.pareto(a, 10000) + x_m)
# distribution_mean = undefined
distribution_median = x_m * (2)**(1/a)
```

```python
def cum_average(x):
    cum_average = []

    for i in range(1,len(x)):
        subset = x[:i]
        cum_average.append(np.mean(subset))

    cum_average = np.array(cum_average)
    return cum_average
def cum_std(x):
    std_average = []

    for i in range(1, len(x)):
        subset = x[:i]
        std_average.append(np.std(subset))

    cum_std = np.array(std_average)
    return std_average

def cum_median(x):
    cumulative_median = []

    for i in range(1, len(x)):
        subset = x[:i]
        cumulative_median.append(np.median(subset))

    cumulative_median = np.array(cumulative_median)
    return cumulative_median

```

```python
pareto_mean = cum_average(s)
pareto_std = cum_std(s)
pareto_median = cum_median(s)
```

```python
x_vals = np.arange(len(pareto_mean))
yerr = pareto_std / np.sqrt(np.arange(1, len(pareto_mean)+1))

plt.plot(x_vals, pareto_mean)


plt.errorbar(
    x_vals,
    pareto_mean,
    yerr=yerr,
    capsize=1,
    fmt="none",        # prevents double plotting
    ecolor="purple"
)

plt.xlabel("Iteration")
plt.ylabel("Cumulative Average")
plt.legend()
plt.show()
```

```python
plt.plot(x_vals, pareto_median)

plt.axhline(y=distribution_median, color="r", linestyle="--", label=f"median = {distribution_median}")

plt.xlabel("Iteration")
plt.ylabel("Cumulative Average")
plt.legend()
plt.show()
```
